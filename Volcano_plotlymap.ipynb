```python
import os
os.environ["MAPBOX_TOKEN"] = 'pk.eyJ1IjoibWF0dGVvMzAyMSIsImEiOiJjbHljbjlwd3Mxc2g0MmxwcXdtdHRpa3U0In0.zFz5Iy1iOMtIyaabtbDCdQ'
```

```python
df_volc = pd.read_csv('volcano_comb_index.csv')
```

```python
fig = px.scatter_mapbox(df_volc,
                        lat='Latitude',
                        lon='Longitude',
                        color='MAX VEI',
                        color_continuous_scale=px.colors.sequential.Plotly3,
                        hover_name='Volcano Name',
                        hover_data=['Primary Volcano Type', 'First Eruption Year',
                                    'Last Eruption Year', 'Eruption Count', 'AVG VEI', 'MAX VEI',
                                    'Total_Deaths', 'Population within 5 km', 'Pop 10km', 'Pop 30km',
                                    'Pop 100 km'],
                        zoom=1.5,
                        height=1000,
                        width=1800)

# Without API KEY
  # fig.update_layout(mapbox_style='carto-positron')

fig.update_layout(mapbox_style='light', mapbox_accesstoken=os.environ["MAPBOX_TOKEN"])
fig.update_traces(marker=dict(size=5))

fig.show()
```

```python
df_index = df_volc.loc[df_volc['composite_index'] > -0.4]
```

```python
fig = px.scatter_mapbox(df_index,
                        lat='Latitude',
                        lon='Longitude',
                        color='Pop 30km',
                        color_continuous_scale=px.colors.sequential.Plotly3,
                        size='Pop 30km',
                        hover_name='Volcano Name',
                        hover_data=['Primary Volcano Type', 'First Eruption Year',
                                    'Last Eruption Year', 'Eruption Count', 'AVG VEI', 'MAX VEI',
                                    'Total_Deaths', 'Population within 5 km', 'Pop 10km', 'Pop 30km',
                                    'Pop 100 km'],
                        zoom=1.5,
                        height=1000,
                        width=1800)

fig.update_layout(mapbox_style='dark', mapbox_accesstoken=os.environ["MAPBOX_TOKEN"])

fig.show()
```

```python
fig = px.density_mapbox(df_index,
                        lat='Latitude',
                        lon='Longitude',
                        z='composite_index',
                        color_continuous_scale=px.colors.sequential.Burg,
                        radius=15,
                        center=dict(lat=df_volc['Latitude'].mean(), lon=df_volc['Longitude'].mean()),
                        hover_name='Volcano Name',
                        hover_data=['Primary Volcano Type', 'First Eruption Year',
                                    'Last Eruption Year', 'Eruption Count', 'AVG VEI', 'MAX VEI',
                                    'Total_Deaths', 'Population within 5 km', 'Pop 10km', 'Pop 30km',
                                    'Pop 100 km'],
                        zoom=1.5,
                        height=1000,
                        width=1800)

fig.update_layout(mapbox_style='dark', mapbox_accesstoken=os.environ["MAPBOX_TOKEN"])

fig.show()
```
